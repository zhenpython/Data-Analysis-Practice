## Project Scenario:

An international firm that is looking to expand its business in different countries across the world has recruited you. You have been hired as a junior Data Engineer and are tasked with creating a script that can extract the list of the top 10 largest economies of the world in descending order of their GDPs in Billion USD (rounded to 2 decimal places), as logged by the International Monetary Fund (IMF).


```python
import numpy as np
import pandas as pd
```

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings('ignore')
```

```python
url="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

```python
tables = pd.read_html(url)
```

```python
df = tables[3]
print(df.shape)
```

```python
# Replace the column headers with column numbers
df.columns = range(df.shape[1])
```

```python
df2 = df[[0, 2]]
df2
```

```python
# Top 10 economies
df2 = df2.iloc[1: 11]
df2
```

```python
df2.columns = ['Country', 'GDP (Million USD)']
df2
```

```python
# Convert million to billion USD
df2['GDP (Million USD)'] = ((df2['GDP (Million USD)']).astype(int) / 1000)
df2['GDP (Million USD)'] = np.round(df2['GDP (Million USD)'], 2)
df2.rename(columns={'GDP (Million USD)': 'GDP (Billion USD)'}, inplace=True)
df2
```

```python
#Load Dataframe to CSV file
df2.to_csv('Largest Economies.csv', index = False)
```
