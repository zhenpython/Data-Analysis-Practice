```python
# Hilangkan baris duplikat
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.describe()
```

```python
avg_capita = df.groupby('Year')['Explained by: Log GDP per capita'].mean().reset_index()
```

```python
plt.plot(avg_capita['Year'], avg_capita['Explained by: Log GDP per capita'], marker='o', linestyle='-')
plt.title('Average Explained by: Log GDP per capita Over Time')
plt.xlabel('Year')
plt.ylabel('Average Value')
plt.grid(True)
plt.show()
```

```python
df_indonesia = df[df['Country name'] == "Indonesia"]
```

```python
plt.plot(df_indonesia['Year'], df_indonesia['Explained by: Log GDP per capita'], marker='o', linestyle='-')
plt.title('Explained by: Log GDP per capita in Indonesia Over Time')
plt.xlabel('Year')
plt.ylabel('Value')
```
